```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataset=pd.read_csv('Algerian_forest_fires_dataset_UPDATE.csv',header=1)
```

```python
dataset.head()
```

```python
#Finding missing Values
dataset[dataset.isnull().any(axis=1)] # OR dataset.isnull().sum()
```

```python
dataset.loc[:122,'Region']=0
dataset.loc[122:,'Region']=1
df=dataset
```

```python
df.info()
```

```python
df[['Region']]=df[['Region']].astype(int)
```

```python
df.info()
```

```python
df[df.isnull().any(axis=1)]
```

```python
df=df.dropna().reset_index(drop=True)
```

```python
df[df.isnull().any(axis=1)]
```

```python
df.isnull().sum()
```

```python
df.iloc[[122]]
```

```python
df=df.drop(122).reset_index(drop=True)
```

```python
df.columns
```

```python
# Removing Spaces from the column names
df.columns=df.columns.str.strip()
df.columns
```

```python
df.info()
```

```python
df.head()
```

### Changing the Datatypes of the Columns

```python
df[['day','month','year','Temperature','RH','Ws']]=df[['day','month','year','Temperature','RH','Ws']].astype(int)
```

```python
df.info()
```

### Changing the other columns to floar datatype

```python
objects=[features for features in df.columns if df[features].dtypes=='O']
```

```python
for i in objects:
    if i!='Classes':
        df[i]=df[i].astype(float)
```

```python
df.info()
```

```python
df.to_csv('Algerian_forest_fires_Cleaned_Dataset.csv',index=False)
```

### Performing EDA

```python
df_copy=df.drop(['day','month','year'],axis=1)
```

```python
df_copy.head()
```

```python
# Encoding the Classes features into 0s and 1s for not fire and fire
df_copy['Classes']=np.where(df_copy['Classes'].str.contains('not fire'),0,1)
```

```python
df_copy[['Classes']].value_counts()
```

```python
df_copy.head()
```

```python
import seaborn as sns
```

```python
#Plotting density plot
plt.style.use('seaborn-v0_8')
df_copy.hist(bins=50, figsize=(20,15))
plt.show()
```

```python
percentage=df_copy[['Classes']].value_counts(normalize=True)
```

```python
percentage
```

```python
classlabel=['Fire','Not Fire']
plt.Figure(figsize=(12,7))
plt.pie(percentage,labels=classlabel,autopct='%1.1f%%')
plt.title("Pie Chart of Classes")
```

```python
df_copy.corr()
```

```python
sns.heatmap(df_copy.corr())
```

```python
df['Classes']=np.where(df['Classes']==0,'not fire','fire')
```

```python
##Monthly Analysis of Fire
dftemp=df.loc[df['Region']==0]
sns.set_style('whitegrid')
sns.countplot(x='month',hue='Classes',data=dftemp)
plt.title("Fire Analysis for Bejaia Region",weight='bold')
plt.xlabel('Month',weight='bold')
plt.ylabel('Count',weight='bold')
```

```python
dftemp=df.loc[df['Region']==1]
sns.set_style('whitegrid')
sns.countplot(x='month',hue='Classes',data=dftemp,palette='plasma')
plt.title("Fire Analysis for Sidi-Bel Abbes Region Region",weight='bold')
plt.xlabel('Month',weight='bold')
plt.ylabel('Count',weight='bold')
```
